### simple line graph

```python
# defining x axis and y axis values
x=[1,3,5,7,9,]
y=[4,6,11,2,9]
plt.plot(x,y)
# you can also insert x and y values directly into the plot function
# plt.plot([1,3,5,7,9,],[4,6,11,2,9])
plt.show()
```

### Plotting sine and cosine function


```python
x = np.arange(0, 6.5, 0.01)
# this arrange function takes the values from 0 all the way to 6.5 with an interval of 0.1
y1 = np.sin(x)
y2 = np.cos(x)
plt.figure(figsize=(8,5), dpi=100)
plt.plot(x, y1, label="sin(x)")
plt.plot(x, y2, label="cos(x)")
plt.title("Sin function graph")
plt.xlabel("x(Angle)")
plt.ylabel("Sin(x)")
plt.legend()
# this legend() function shows the labels on for the graphs
plt.show()
```

### Plotting tangent function

```python
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

# define x-axis range
x= np.linspace(-2*np.pi, 2*np.pi, 1000)
# calculate tan(x)
y=np.tan(x)

# since the value of tan(90) approaches infinity, you have the handle the asymptotes
y[np.abs(y)>100]=np.nan

#Create the plot
plt.plot(x,y, label="tan(x)")
# set up the graph accessories
plt.grid(True, linestyle="--", alpha=0.7)
plt.legend(loc="upper right")
plt.title("Plot of the tangent function")
plt.xlabel("x(radian)")
plt.ylabel("tan(x)")

#showing the main axis
plt.axhline(0, color="black", linewidth=0.7)
plt.axvline(0, color="black", linewidth=0.7)
# set x axis ticks and labels
plt.xticks(
    ticks=np.arange(-2*np.pi, 2.5*np.pi, 0.5*np.pi),
    labels= [r"$-2\pi$", r"$-\frac{3\pi}{2}$", r"$-\pi$",r"$-\frac{\pi}{2}$",r"$0$",
             r"$\frac{\pi}{2}$",r"$\pi$",r"$\frac{3\pi}{2}$", r"$-2\pi$" ]
)

# set y axis limit
plt.ylim(-10, 10)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(19680801)


fig, ax = plt.subplots()
for color in ['tab:blue', 'tab:orange', 'tab:green']:
    n = 750
    x, y = np.random.rand(2, n)
    scale = 200.0 * np.random.rand(n)
    ax.scatter(x, y, c=color, s=scale, label=color,
               alpha=0.3, edgecolors='none')

ax.legend()
ax.grid(True)

plt.show()
```

```python
N = 450
x, y = np.random.rand(2, N)
c = np.random.randint(1, 5, size=N)
s = np.random.randint(10, 220, size=N)

fig, ax = plt.subplots()

scatter = ax.scatter(x, y, c=c, s=s)

# produce a legend with the unique colors from the scatter
legend1 = ax.legend(*scatter.legend_elements(),
                    loc="lower left", title="Classes")
ax.add_artist(legend1)

# produce a legend with a cross-section of sizes from the scatter
handles, labels = scatter.legend_elements(prop="sizes", alpha=0.6)
legend2 = ax.legend(handles, labels, loc="upper right", title="Sizes")

plt.show()
```
